El objetivo de este notebook  es construir un modelo de probabilidad para describir las reclamaciones agregadas para un sistema de seguros que ocurren en un período de tiempo fijo. El *sistema de seguros* podría ser una póliza única, un contrato de seguro grupal, una línea comercial . En este notebook, las reclamaciones agregadas se refieren al número o la cantidad de reclamaciones de una cartera de contratos de seguro. Sin embargo, el marco de modelado se puede aplicar fácilmente en la configuración más general.

Considere una cartera de seguros de $n$ contratos individuales, y 
$S$ denotara las pérdidas agregadas de la cartera en un período de tiempo determinado. Hay dos enfoques para modelar las pérdidas agregadas
S, el **modelo de riesgo individual** y el **modelo de riesgo colectivo**. El modelo de riesgo individual enfatiza la pérdida de cada contrato individual y representa las pérdidas agregadas como:

\begin{align}  
S=X1+X2+⋯+Xn
\end{align}.
 

dónde
$X_i(i=1,...,n)$ se interpreta como el monto de la pérdida del
i-th contrato. Cabe destacar que $n$ denota el número de contratos en la cartera y, por lo tanto, es un **número fijo** en lugar de una variable aleatoria. Para el modelo de riesgo individual, generalmente se asume $X_i 's$ Son independientes. Debido a las diferentes características del contrato, como la cobertura y la exposición, el
$X_i$ No se distribuyen necesariamente de forma idéntica. Una característica notable de la distribución de cada
$X_i$ es la masa de probabilidad en cero correspondiente al evento de no reclamaciones.

El modelo de riesgo colectivo representa las pérdidas agregadas en términos de una distribución de frecuencia y una distribución de gravedad:
\begin{align}.
S = X_1+ X_2 +⋯+ X_n
\end{align}.
 

Aquí, uno piensa en un número aleatorio de afirmaciones.
norte que puede representar el número de pérdidas o el número de pagos. Por el contrario, en el modelo de riesgo individual, utilizamos un número fijo de contratos n. Pensamos en
$X_1,X_2,...,X_n$como representando el monto de cada pérdida. Cada pérdida puede corresponder o no a un contrato único. Por ejemplo, puede haber múltiples reclamaciones derivadas de un solo contrato. Es natural pensar enc $X>0$ porque si $X_i$=0
  entonces no se ha producido ningún reclamo. Normalmente asumimos que condicional a $N=n$ $,X1,X2,...,X_N$ son iid variables aleatorias. La distribución de
$N$ se conoce como la distribución de frecuencia, y la distribución común de $X$
  se conoce como distribución de severidad. Además asumimos $n$ y $X$ son **independientes**. Con el modelo de riesgo colectivo, podemos descomponer las pérdidas agregadas en la frecuencia (
n) proceso y la severidad (X) modelo. Esta flexibilidad le permite al analista comentar sobre estos dos componentes separados. Por ejemplo, el crecimiento de las ventas debido a estándares de suscripción más bajos podría conducir a una mayor frecuencia de pérdidas, pero podría no afectar la gravedad. De manera similar, la inflación u otras fuerzas económicas podrían tener un impacto en la severidad pero no en la frecuencia.

## Enfoque desde el Riesgo individual

Bajo el enfoque del "Riesgo individual" , como se menciono, se considera a $X_i$ como la perdida asociada al $i$-esimo contrato y por lo tanto , la perdida agregada de un grupo de contratos en un portafolio se consideraria del siguiente modo :

\begin{align}  
S=X1+X2+⋯+Xn
\end{align}.

Respecto a los momentos de esta *perdida agregada*, no necesariamente se asume que las variables aleatorias que la conforman sean identicamente distribuidas por lo que tendremos 


\begin{align}  
E(S_n)=\sum_{i=1}^n E(Xi)
\end{align}.
 
.
 

Pero el problema sigue vigente, esto es, modelar la **distribucion de la perdida** agregada una forma asumiendo un cantidad($n$) significativa de contratos como parte del portafolio es mediante el teorema de limite central , asi uno podria aproximar la distribucion de $S_n$ para un $n$ grande .

Entonces denotando  $μ(S_n) =E(S_n)$ y  $σ^2(Sn) =Var(Sn)$ y sea  
$Z∼N(0,1)$, a standard normal random variable with cdf  $Φ$
. Then the cdf of  $S_n$
  puede ser aproximada de la siguiente manera:
\begin{align}
FSn(s)=Pr(Sn≤s)=Pr(Sn−μ(Sn)/σ(S_n)≤s−μ(S_n)/σ(S_n)≈Pr(Z≤s−μS_nσS_n)=Φ(s−μS_n/σS_n)
\end{align}
